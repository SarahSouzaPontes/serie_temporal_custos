```python
import pandas as pd
import matplotlib.pyplot as plt

# Exemplo de dados (substitua isso pelos seus próprios dados)
data = {
    "Data": ["2023-01-01", "2023-02-01", "2023-03-01"],
    "Usuários": [100, 150, 200],
    "Custos": [500, 600, 750]
}

# Criar DataFrame com os dados
df = pd.DataFrame(data)
df["Data"] = pd.to_datetime(df["Data"])

# Criar gráfico de série temporal
plt.figure(figsize=(10, 6))
plt.plot(df["Data"], df["Usuários"], label="Usuários", marker='o')
plt.plot(df["Data"], df["Custos"], label="Custos", marker='o')
plt.xlabel("Data")
plt.ylabel("Quantidade")
plt.title("Crescimento de Usuários e Custos ao Longo do Tempo")
plt.legend()
plt.xticks(rotation=45)
plt.tight_layout()

plt.show()
```
